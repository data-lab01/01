```python
#importing required modules

import plotly as py
import plotly.graph_objs as go
import ipywidgets as widgets 
import pandas as pd
import plotly.express as px

py.offline.init_notebook_mode(connected=True)
```

```python
#opening up my Csv grants file using panda

df = pd.read_csv('grants.csv')

df
```

```python
#filtering out only data where grantee City is New york

newyork = df[df.GranteeCity == "New York"]
```

```python
#Getting data for my x and y axis 
newyork_xaxis = newyork.AwardYear
newyork_yaxis = newyork.Total_Active_Grant_Financial_Assistance
```

```python
#drawing my graph

graph  = go.Scatter(
    x = newyork_xaxis,
    y = newyork_yaxis,
    
)

fig = go.Figure(data=graph)

py.offline.iplot(fig)
```
